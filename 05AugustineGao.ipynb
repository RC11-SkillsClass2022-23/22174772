```python
import json
import os
from moviepy.editor import *
from moviepy import *
```

```python
def loadVideosToInfile(folder):
    videos = os.listdir(folder)
    with open('input', 'w') as inputfile:
      for v in videos:
        if v.split('.')[-1] == 'mp4':
          inputfile.write(v+'\n')
```

```python
import numpy as np
```

```python
sFeatures = []
for s in segments[0]['clips']:
  sFeatures.append(np.array(s['features'])) 
```

```python
print(len(sFeatures))
```

```python
print(len(sFeatures[0]))
```

```python
longFeatures = []
for s in segments[1]['clips']:
  longFeatures.append(np.array(s['features'])) 
```

```python
print(len(longFeatures))
```

```python
bestRangeStart = 0
bestRangeScore = 100000000
for i in range(len(longFeatures)-len(sFeatures)):
  distance = 0
  for j in range(len(sFeatures)):
    d = np.linalg.norm(longFeatures[i+j]-sFeatures[j])
    distance += d
  print(distance)
  if distance < bestRangeScore:
    bestRangeScore = distance
    bestRangeStart = i
```

```python
bestRangeStart
```

```python
catVideo = VideoFileClip(r"/content/drive/MyDrive/RC11_Skill_Classes/cat01/long01.mp4")
catfps = catVideo.fps
catFrameLength = 1/catfps

newFrames = len(longFeatures)*16
startFrame = bestRangeStart*16

startTime = startFrame*catFrameLength
endTime = startTime+newFrames*catFrameLength

catMatchedMhVideo = catVideo.subclip(startTime,endTime)
```
