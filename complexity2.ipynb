```python
#1.1

import numpy as np
def SMM(A,B):
  n = len(A)
  C = np.zeros((n,n))
  for i in range(1,n+1):
    for j in range(1,n+1):
      C[i-1][j-1] = 0
      for k in range(1,n+1):
        C[i-1][j-1] += int(A[i-1][k-1]) * int(B[k-1][j-1])
     
  return C
```

```python
A = [[1,2],[3,4]]
B = [[1,2],[3,4]]
SMM(A,B)
```
